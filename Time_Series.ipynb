### Step 1 - Read the AirPassengers Data

```python
import pandas as pd
url = "https://raw.githubusercontent.com/utkarshg1/Neural-Net-Datasets/refs/heads/main/AirPassengers.csv"
df = pd.read_csv(url, index_col="Month", parse_dates=True)
df.head()
```

### Visualize the time series data

```python
df["Passengers"].plot(kind="line")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12, 5))
sns.lineplot(data=df)
plt.title("Air Passengers")
plt.show()
```

image.png

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
Q = seasonal_decompose(df["Passengers"], model="multiplicative")
```

```python
df["Passengers"].plot()
Q.trend.plot()
```

```python
Q.seasonal.plot()
```

```python
Q.resid.plot()
```

### Time series forecasting
ARIMA - Auto Regressive Integrated Moving Average

### Find the airline passengers for next 4 years = 4*12 = 48 month

```python
df.tail()
```

```python
from pmdarima import auto_arima
```

```python
model = auto_arima(df["Passengers"], seasonal=True, m=12)
```

```python
model.order
```

### ARMIA(2, 1, 1)

```python
forecast, conf = model.predict(n_periods=48, return_conf_int=True)
```

```python
forecast
```

```python
conf
```

```python
res = forecast.to_frame()
res.columns = ["Forecast"]
res[["Lower", "Upper"]] = conf
res = res.round(0)
res
```

```python
res.to_csv("ARIMARes.csv")
```

### Visualizing with conf interval

```python
plt.figure(figsize=(12, 5))
plt.plot(df["Passengers"])
plt.plot(res.index, res["Forecast"], color="red")
plt.fill_between(res.index, res["Lower"], res["Upper"], color="pink")
plt.show()
```
