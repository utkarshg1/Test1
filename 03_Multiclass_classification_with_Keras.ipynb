# Step 3 - Seperate X and Y(Weather Type)

# Step 4 - Train Test Split

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(
    X, Y, test_size=0.2, random_state=42
)
```

# Step 5 - Apply Preprocessing on X

```python
X.dtypes
```

```python
X.select_dtypes(include="object").nunique()
```

```python
num_cols = X.select_dtypes(include="number").columns.tolist()
cat_cols = X.select_dtypes(include="object").columns.tolist()
```

```python
num_cols
```

```python
cat_cols
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipe = make_pipeline(
    SimpleImputer(strategy = "median"),
    StandardScaler()
)
```

```python
cat_pipe = make_pipeline(
    SimpleImputer(strategy = "most_frequent"),
    OneHotEncoder(handle_unknown="ignore", sparse_output=False, drop="first")
)
```

```python
pre = ColumnTransformer(
    [
        ("num", num_pipe, num_cols),
        ("cat", cat_pipe, cat_cols)
    ]
).set_output(transform="pandas")
```

```python
pre.fit(xtrain)
```

```python
xtrain_pre = pre.transform(xtrain)
xtest_pre = pre.transform(xtest)
```

```python
xtrain_pre.head()
```

```python
xtest_pre.head()
```

# OneHotEncode target feature

```python
df["Weather Type"].unique()
```

```python
ytrain_pre = pd.get_dummies(ytrain, dtype=int)
ytrain_pre.head()
```

```python
ytest_pre = pd.get_dummies(ytest, dtype=int)
ytest_pre.head()
```

# Step 6 - Train the neural net

image.png

```python
xtrain_pre.shape[1]
```

```python
ytrain_pre.shape[1]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense, Dropout
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping
```

```python
model = Sequential(
    [
        Input(shape=(xtrain_pre.shape[1],)),
        Dense(32, activation="relu", kernel_regularizer=l2(0.01)),
        Dropout(0.3),
        Dense(16, activation="relu", kernel_regularizer=l2(0.01)),
        Dropout(0.2),
        Dense(ytrain_pre.shape[1], activation="softmax")
    ]
)
```

```python
model.summary()
```

```python
model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
es = EarlyStopping(patience=10, restore_best_weights=True)
```

```python
hist = model.fit(
    xtrain_pre, ytrain_pre, validation_split=0.2, epochs=100, callbacks = [es]
)
```

# Step 7 - Plot learning curve

```python
import matplotlib.pyplot as plt

plt.plot(hist.history["loss"], label="Train Loss")
plt.plot(hist.history["val_loss"], label="Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Categorical Cross entropy")
plt.title("Learning Curve")
plt.show()
```

# Model Evaluation

```python
train_loss, train_acc = model.evaluate(xtrain_pre, ytrain_pre)
print(f"Train Loss : {train_loss:.2f}")
print(f"Train Accuracy : {train_acc:.2f}")
```

```python
test_loss, test_acc = model.evaluate(xtest_pre, ytest_pre)
print(f"Test Loss : {test_loss:.2f}")
print(f"Test Accuracy : {test_acc:.2f}")
```
