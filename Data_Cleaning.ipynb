```python
from warnings import filterwarnings
filterwarnings("ignore")
```

```python
import pandas as pd
df = pd.read_csv("http://bit.ly/SampleDataset")
df.head()
```

```python
df.columns
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns
```

```python
df.columns = [col.replace(" ", "_") for col in df.columns]
```

```python
df.columns
```

```python
df.columns = (df.columns
  .str.strip()
  .str.lower()
  .str.replace(" ", "_")
  .str.replace("(", "")
  .str.replace(")","")
)
```

```python
df.head()
```

```python
df["city"] = df["city"].str.lower()
```

```python
df.head()
```

```python
df["height"] = df["height"]*2.54
```

```python
df.dtypes
```

```python
df["height"] = df["height"].astype(int)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df["country"].unique()
```

```python
df["country"] = (
    df["country"]
    .str.strip()
    .str.lower()
    .str.replace("germanyy", "germany")
)
```

```python
df.head()
```

```python
df["country"].unique()
```

```python
df["city"].unique()
```

```python
df["city"] = df["city"].str.strip()
```

```python
df["city"].unique()
```

```python
df["name"].unique()
```

```python
df["name"] = df["name"].str.strip()
```

```python
df["name"].unique()
```

```python
df["total_account"] = df["account_a"] + df["account_b"]
```

```python
df.head()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates(keep="first").reset_index(drop=True)
```

```python
avg_weight = df["weight"].mean()
df["weight"] = df["weight"].fillna(avg_weight)
```

```python
df["account_a"] = df["account_a"].fillna(0)
```
