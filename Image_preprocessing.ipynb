# Image preprocessing
1. Colored image 3 channels -> Red, Green, Blue
2. Grayscale - 1 channel
3. Pixel range is between 0-255

```python
image_path = "/content/Alphabet Example.jpg"
print(image_path)
```

```python
import cv2
img1 = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
img1
```

```python
import matplotlib.pyplot as plt

plt.imshow(img1, cmap="gray")
```

# Visualizing pixels

```python
import seaborn as sns

plt.figure(figsize=(15, 15))
sns.heatmap(img1, annot=True, fmt="d")
plt.show()
```
