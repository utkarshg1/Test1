## Data Preprocessing
Pixel 0 - 255

```python
xtrain_scl = xtrain.astype("float32")/255
xtest_scl = xtest.astype("float32")/255
```

```python
xtrain_scl[0]
```

```python
xtrain_scl.shape
```

```python
xtrain_scl = np.expand_dims(xtrain_scl, -1)
xtest_scl = np.expand_dims(xtest_scl, -1)
```

```python
xtrain_scl.shape
```

```python
xtest_scl.shape
```

```python
np.unique(ytrain)
```

```python
from keras.utils import to_categorical

ytrain_cat = to_categorical(ytrain)
ytest_cat = to_categorical(ytest)
```

```python
ytrain_cat[0:5]
```

```python
ytrain[0:5]
```

# Build the Convolutional Neural Network

image.png

```python
from keras.models import Sequential
from keras.layers import Input, Convolution2D, MaxPool2D, Flatten, Dropout, Dense
```

```python
model = Sequential(
    [
        Input(shape=(28, 28, 1)),

        Convolution2D(32, kernel_size=(3, 3), activation="relu"),
        MaxPool2D(pool_size=(2, 2)),

        Convolution2D(64, kernel_size=(3, 3), activation="relu"),
        MaxPool2D(pool_size=(2, 2)),

        Flatten(),

        Dropout(0.5),

        Dense(10, activation="softmax")
    ]
)
```

```python
model.summary()
```

```python
model.compile(loss="categorical_crossentropy", optimizer="adam", metrics = ["accuracy"])
```

```python
hist = model.fit(xtrain_scl, ytrain_cat, validation_split=0.1, batch_size=128, epochs=20)
```

### Plot the learning curve

```python
plt.figure(figsize=(8, 5))
plt.plot(hist.history["loss"], label="Training Loss")
plt.plot(hist.history["val_loss"], label="Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Categorical Crossentropy")
plt.title("Learning Curve")
plt.legend()
plt.show()
```

### Model Evaluatation

```python
model.evaluate(xtrain_scl, ytrain_cat)
```

```python
model.evaluate(xtest_scl, ytest_cat)
```

### Confusion Matrix

```python
yprob_train = model.predict(xtrain_scl)
yprob_train[0:5]
```

```python
ypred_train = np.argmax(yprob_train, axis=1)
```

```python
ypred_train[0:5]
```

```python
yprob_test = model.predict(xtest_scl)
ypred_test = np.argmax(yprob_test, axis=1)
ypred_test[0:5]
```

```python
ytest[0:5]
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_predictions(ytest, ypred_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(ytest, ypred_test))
```

```python
def predict_digit(model, img_path):
  img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
  plt.imshow(img, cmap = "gray")
  plt.show()
  img = img.astype("float32")/255
  img = img.reshape(1, 28, 28, 1)
  prob = model.predict(img)
  max_prob = np.max(prob)
  pred = np.argmax(prob, axis=1)
  return pred, max_prob
```
