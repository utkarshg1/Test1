### Do preprocessing on x and y
1. pixel range is between 0-255

```python
xtrain_scl = xtrain.astype("float32")/255
xtest_scl = xtest.astype("float32")/255
```

```python
xtrain_scl[0]
```

```python
xtrain_scl = np.expand_dims(xtrain_scl, -1)
xtest_scl = np.expand_dims(xtest_scl, -1)
```

```python
xtrain_scl.shape
```

```python
xtest_scl.shape
```

### For multiclass classification convert y to onehot encoding

```python
from keras.utils import to_categorical

ytrain_pre = to_categorical(ytrain)
ytest_pre = to_categorical(ytest)
```

### Start building the CNN
image.png

```python
xtrain.shape
```

```python
from keras.models import Sequential
from keras.layers import Input, Dense, Dropout, Conv2D, MaxPooling2D, Flatten
```

```python
model = Sequential(
    [
        Input(shape=(28, 28, 1)),
        Conv2D(filters=32, kernel_size=(3, 3), activation="relu"),
        MaxPooling2D(pool_size=(2,2)),
        Conv2D(filters=64, kernel_size=(3,3), activation="relu"),
        MaxPooling2D(pool_size=(2, 2)),
        Flatten(),
        Dropout(0.5),
        Dense(units=10, activation="softmax")
    ]
)
```

```python
model.summary()
```

```python
model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
hist = model.fit(xtrain_scl, ytrain_pre, validation_split=0.1, epochs=20, batch_size=128)
```

### Plot the learning curve

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.plot(hist.history["loss"], label="Train Loss")
plt.plot(hist.history["val_loss"], label="Val Loss")
plt.xlabel("Epochs")
plt.ylabel("Categorical Crossentropy")
plt.title("Learning Curve")
plt.legend()
plt.show()
```

```python
def predict_label(model, x):
  yprob = model.predict(x)
  ypred = yprob.argmax(axis=1)
  max_probs = yprob.max(axis=1)
  return ypred, max_probs
```

```python
def predict_image(model, path:str):
  # Read image in numpy format
  xnew = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

  # Plot the image
  plt.imshow(xnew, cmap="gray")
  plt.show()

  # Preprocess x data
  xnew_scl = xnew.astype("float32")/255
  xnew_scl = xnew_scl.reshape(1, 28, 28, -1)

  # Provide prediction with probability
  probs = model.predict(xnew_scl)
  preds = probs.argmax()
  max_prob = probs.max()

  return preds, max_prob
```
